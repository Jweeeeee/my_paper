## 2 相关理论概述

### 2.1 词向量

词向量是文本分析的基本单位，是单词的数据表现形式。词向量有多种呈现方式，如离散形式，独热编码，词共现矩阵等方式。

本文的词向量主要呈现的信息是文章各词语的频次。本文对培养方案文本的分词均以调用jieba等库实现。

### 2.2 TF-IDF算法

TF-IDF（term frequency–inverse document frequency）是一种用于数据挖掘和搜索引擎的常用加权技术，用以评估单个词语对于一个文本集合或其中一份文本的重要程度。词语的重要性随着它在文件中出现的次数成比例增加，但同时会随着它在文本集合中出现的频率成反比下降。TF-IDF值由𝑇 𝐹 × 𝐼𝐷𝐹计算得出。

TF是词频(Term Frequency)，表示单个词语在一篇文章中出现的频率，用以度量单个词语在一篇文章中的重要性。

IDF是逆向文本频率指数(Inverse Document Frequency)，可以表示单个词语在一些文章中的普遍重要性。

### 2.3 余弦相似度

在文本分析中，每个词项被赋予不同的维度，而一个维度由一个向量表示，其各个维度上的值对应于该词项在文档中出现的频率。余弦相似度可以给出两篇文档在其主题方面的相似度。

培养方案文本分词结果以及词云生成

```python
# 文本预处理
pattern = re.compile(u'\t|\n|\.|-|:|;|\)|\(|\?|"') # 定义正则表达式匹配模式
string_data = re.sub(pattern, '', txts_jb) # 将符合模式的字符去除

# 文本分词
seg_list_exact = jb.cut(string_data, cut_all = False) # 精确模式分词
object_list = []
remove_words = [u'的', u'，',u'和',u'与', u'是', u'随着', u'对于',u'为', u'对',u'等',u'能',u'都',u'。',u' ',u'、',u'中',u'在',u'了',
                u'通常',u'如果',u'我们',u'需要',u'√',u'\u3000',u'\u200b',u'0',u'1',u'2',u'3',u'4',u'5',u'6',u'7',u'8',u'9',
                u'一',u'二',u'三',u'四',u'五',u'：',u'；',u'）',u'（',u'《',u'》',u'/'] # 自定义去除词库

for word in seg_list_exact: # 循环读出每个分词
    if word not in remove_words: # 如果不在去除词库中
        object_list.append(word) # 分词追加到列表

# 词频统计
word_counts = col.Counter(object_list) # 对分词做词频统计
word_counts_top50 = word_counts.most_common(50) # 获取前50最高频的词
print(word_counts_top50) # 输出检查

# 词频展示
mask = np.array(Image.open(r'F:\JWE\图片\壁纸\292773.jpg')) # 定义词频背景
wc = wd.WordCloud(
    font_path=r'C:\Users\Administrator\Desktop\论文数据\simhei.ttf', # 设置字体格式
    mask=mask, # 设置背景图
    max_words=100, # 最多显示词数
    max_font_size=150 # 字体最大值
)

wc.generate_from_frequencies(word_counts) # 从字典生成词云
image_colors = wd.ImageColorGenerator(mask) # 从背景图建立颜色方案
wc.recolor(color_func=image_colors) # 将词云颜色设置为背景图方案
plt.imshow(wc) # 显示词云

plt.axis('off') # 关闭坐标轴
plt.show() # 显示图像
```

```python
def text_clean(txt):
    # 文本预处理
    pattern = re.compile(u'\t|\n|\.|-|:|;|\)|\(|\?|"') # 定义正则表达式匹配模式
    string_data = re.sub(pattern, '', txt) # 将符合模式的字符去除

    # 文本分词
    seg_list_exact = jb.cut(string_data, cut_all = False) # 精确模式分词
    object_list = []
    remove_words = [u'的', u'，',u'和',u'与', u'是', u'随着', u'对于',u'为', u'对',u'等',u'能',u'都',u'。',u' ',u'、',u'中',u'在',u'了',
                    u'通常',u'如果',u'我们',u'需要',u'√',u'\u3000',u'\u200b',u'0',u'1',u'2',u'3',u'4',u'5',u'6',u'7',u'8',u'9',
                    u'一',u'二',u'三',u'四',u'五',u'：',u'；',u'）',u'（',u'《',u'》',u'/'] # 自定义去除词库

    for word in seg_list_exact: # 循环读出每个分词
        if word not in remove_words: # 如果不在去除词库中
            object_list.append(word) # 分词追加到列表
    return object_list
```
